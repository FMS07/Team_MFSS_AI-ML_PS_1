```python
import os
import http.client
import json
import PyPDF2
from groq import Groq

# Set your API key for the Groq client here
GROQ_API_KEY = "g7"  # Replace with your actual Groq API key
JOOBLE_API_KEY = ""  # Replace with your actual Jooble API key
JOOBLE_HOST = ""


def extract_text_from_pdf(pdf_path):
    """
    Extracts text from a PDF file.
    """
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
    return text


def analyze_resume_with_llama(text):
    """
    Analyze the extracted text using the Groq API.
    """
    client = Groq(api_key=GROQ_API_KEY)

    # Prompt for analysis
    prompt = f"""
    Analyze the following resume text and identify key domains related to the skills, projects, internships, and interests mentioned.
    Limit the identified domains to a maximum of 10 to 15 most important ones. Provide only the domain names without any numbers, descriptions, or attached explanations.

    Resume Text: {text}
    """
    completion = client.chat.completions.create(
        model="llama3-70b-8192",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7,
        max_tokens=512,
        top_p=1,
        stream=True,
        stop=None,
    )

    # Gather output
    analysis_result = ""
    for chunk in completion:
        chunk_content = chunk.choices[0].delta.content or ""
        analysis_result += chunk_content

    return analysis_result


def extract_and_limit_domains(analysis_result, max_domains=15):
    """
    Extract important domains from the analysis result and limit to a maximum count.
    """
    domains = []
    lines = analysis_result.split("\n")

    # Extract domains and clean them
    for line in lines:
        domain = line.strip()  # Strip leading/trailing spaces
        if domain and len(domain.split()) <= 5:  # Ensure concise domain names (up to 5 words)
            domains.append(domain)

    # Limit the number of domains
    return domains[:max_domains]


def get_job_listings_from_jooble(keywords, location="", max_results=2):
    """
    Get limited job listings from Jooble API using keywords and location.
    """
    connection = http.client.HTTPSConnection(JOOBLE_HOST)
    headers = {"Content-type": "application/json"}

    # Prepare the body of the request
    body = json.dumps({"keywords": keywords, "location": location})

    # Make the API request
    connection.request("POST", f"/api/{JOOBLE_API_KEY}", body, headers)
    response = connection.getresponse()

    # Process the response
    if response.status == 200:
        data = response.read()
        job_data = json.loads(data)
        # Limit the number of jobs returned
        return job_data.get("jobs", [])[:max_results]
    else:
        print(f"Error: {response.status} {response.reason}")
        return None


def main():
    """
    Main function to handle PDF processing, domain extraction, and job search.
    """
    pdf_path = "/content/Shreyank_Srinivasan_Harisha_Resume_.pdf"  # Replace with the actual path to your PDF

    if not os.path.exists(pdf_path):
        print("Error: The specified file does not exist.")
        return

    # Extract text from the resume
    resume_text = extract_text_from_pdf(pdf_path)

    print("\nAnalyzing Resume...\n")
    analysis = analyze_resume_with_llama(resume_text)

    # Extract and limit domains
    important_domains = extract_and_limit_domains(analysis, max_domains=15)

    print("\nImportant Domains Identified:")
    for domain in important_domains:
        print(domain)

    print("\nSearching for Job Listings...\n")
    for domain in important_domains:
        job_listings = get_job_listings_from_jooble(domain, max_results=2)
        if job_listings:
            print(f"Jobs for domain '{domain}':")
            for job in job_listings:
                print(f"- {job['title']} at {job['company']}")
                print(f"  Location: {job['location']}")
                print(f"  Link: {job['link']}\n")
        else:
            print(f"No jobs found for domain '{domain}'.")


if __name__ == "__main__":
    main()
```
