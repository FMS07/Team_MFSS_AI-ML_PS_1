```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def scrape_public_linkedin_jobs():
    # Set up WebDriver
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(options=options)

    try:
        # Navigate to LinkedIn job search page (public access)
        driver.get("https://www.linkedin.com/jobs/")

        # Wait for the page to load and display job cards
        job_cards_xpath = '/html/body/div[6]/div[3]/div[4]/div/div/main/div/div[2]/div[1]/div/ul/li'
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, job_cards_xpath))
        )

        # Get all job cards dynamically
        job_cards = driver.find_elements(By.XPATH, job_cards_xpath)
        print(f"Total job cards found: {len(job_cards)}")

        # Collect job data
        job_list = []
        for i in range(1, len(job_cards) + 1):
            try:
                # Construct dynamic XPath for the title
                title_xpath = f'{job_cards_xpath}[{i}]/div/div/div[1]/div/div[2]/div[1]/a/span[1]/strong'
                company_xpath = f'{job_cards_xpath}[{i}]/div/div/div[1]/div/div[2]/div[2]/a'
                location_xpath = f'{job_cards_xpath}[{i}]/div/div/div[1]/div/div[2]/div[3]/ul/li[1]'

                title = driver.find_element(By.XPATH, title_xpath).text
                company = driver.find_element(By.XPATH, company_xpath).text
                location = driver.find_element(By.XPATH, location_xpath).text

                job_list.append({
                    "Title": title,
                    "Company": company,
                    "Location": location
                })
            except Exception as e:
                print(f"Error extracting details for job {i}: {e}")

        # Print the collected job data
        for idx, job in enumerate(job_list, start=1):
            print(f"Job {idx}:")
            print(f"  Title: {job['Title']}")
            print(f"  Company: {job['Company']}")
            print(f"  Location: {job['Location']}")
            print("-" * 40)

    finally:
        # Close the browser
        driver.quit()

# Example usage
scrape_public_linkedin_jobs()
```
